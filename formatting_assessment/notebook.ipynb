```python
class ConfusionMatrix:
    def __init__(self, data) -> None:
        self.table = data
    
    @property
    def table(self):
        return self.table
    
    @table.setter
    def table(self, data: dict):
        self.table = pd.DataFrame(
            data=data.get('cfm'),
            columns=data.get('labels'),
            index=data.get('labels')
        )

    def add_producers(self, data):
        self.table.reindex(columns=self.table.columns.tolist() + ['Producers'])
        data = list(map(lambda x: round(x * 100, 2), data))
        self.table['Producers'] = data
        return self
    

cmt = ConfusionMatrix(data)


cmt.add_producers(data=data.get('pro'))
```
