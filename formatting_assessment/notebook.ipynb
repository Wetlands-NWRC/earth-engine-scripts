```python
import json
import pandas as pd
import numpy as np
```

```python
with open("./aoi_novascotia_output_20230815_104538_122_assessment_assessment-122.geojson", 'r') as obj:
    data = json.load(obj)
features = data['features']
props = [_.get("properties") for _ in features]
data = {k: v for _ in props for k, v in _.items()}
```

```python
data
```

```python
# Confusion Matrix
cfm = pd.DataFrame(
    data=data.get('cfm'),
    columns=data.get('labels'),
    index=data.get('labels')
)

cfm
```

```python
cfm = cfm.reindex(columns=cfm.columns.tolist() + ['Producers'])

new_index = pd.Index(cfm.index.tolist() + ['Consumers'])
cfm = cfm.reindex(new_index).fillna(value=np.nan)
```

```python
cfm
```

```python
pro = data.get('pro')
pro.append(np.nan)
cfm['Producers'] = pro
cfm
```
