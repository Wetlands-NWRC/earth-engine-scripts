```python
import pandas as pd
from itertools import combinations
```

```python

df = pd.read_csv('./data/raw/samples-124.csv')
df.head()
```

```python
cols_2_keep = [i for i in df.columns if i not in ['.geo', 'system:index', 'isTraining', 'class_name']]
cols_2_keep.insert(0, 'class_name')
cols_2_keep
```

```python
df = df[cols_2_keep]
df.head()
```

```python
labels = df['class_name'].unique().tolist()
combos = combinations(labels, 2)
```

```python
data_frames = []
for combo in combinations(labels, 2):
    if combo[0] != 1:
        break
    print(combo)
    dfin = df.copy()
    data = {'label':[], 'predictor':[], 'score':[]}
    for col in dfin.columns:
        dfin = df[(df['class_name'] == combo[0]) | (df['class_name'] == combo[1])]
        if col == 'class_name':
            continue
        dfc1 = dfin[dfin['class_name'] == combo[0]][col]
        dfc2 = dfin[dfin['class_name'] == combo[1]][col]
        c1_mean = dfc1.mean()
        c2_mean = dfc2.mean()
        cat = pd.concat([dfc1, dfc2])
        std = cat.std()
        score = abs(c1_mean - c2_mean) / std
        print(f'{col}: {score}')
        data['label'].append(f"{combo[0]}:{combo[1]}")
        data['predictor'].append(col)
        data['score'].append(score)
    df_out = pd.DataFrame(data)
    data_frames.append(df_out)
    data = {}

    
```

```python
# df_test = data_frames[0].sort_values(by='score', ascending=False).reset_index(drop=True)
# df_test['rank'] = list(range(1, len(df_test) + 1))
# df_test
```

```python
ranked = []
for df in data_frames:
    df = df.sort_values(by='score', ascending=False).reset_index(drop=True)
    df['rank'] = list(range(1, len(df) + 1))
    ranked.append(df)
```

```python
pd.concat(ranked).to_csv('./data/processed/feature_ranking.csv', index=False)
```
