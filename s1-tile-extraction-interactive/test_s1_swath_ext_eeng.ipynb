```python
import geopandas as gpd
import pandas as pd
```

```python
gdf = gpd.read_file("./data/s1_dv_2019_04_01_2019_10_31.shp", driver="ESRI Shapefile")
gdf.head()
```

```python
from eeng.client.gdftools import extract_s1_swaths

swaths = extract_s1_swaths(gdf=gdf, relobits=[150, 4, 77, 106])
```

```python
swaths
```

```python
swaths['date'] = pd.to_datetime(swaths['date']).dt.strftime('%Y-%m-%d')
swaths.to_file('data/output_swaths.shp')
```
