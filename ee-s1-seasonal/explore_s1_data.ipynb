```python
import pandas as pd
import geopandas as gpd
```

```python
# Read in the data
df = gpd.read_file('data/raw/s1.shp', driver='ESRI Shapefile')
df.head()
```

```python
rel_orbits = df['relativeOr'].unique()
rel_orbits
```

```python
df_106 = df[df['relativeOr'] == 106]
df_106
```

```python
import os

if not os.path.exists('data/processed'):
    os.makedirs('data/processed')

df_106.to_file('data/processed/s1_106.shp', driver='ESRI Shapefile')
```

```python
dates = df_106['date'].unique()
dates
```

```python
swath_dates = {}
for idx, date in enumerate(dates):
    swath_dates[idx] = df_106[df_106['date'] == date]['system_id'].tolist()
swath_dates
```

```python
# batch

if not os.path.exists('data/processed/pairings'):
    os.makedirs('data/processed/pairings')

swaths = {}
for swath in df['relativeOr'].unique():
    df_swath = df[df['relativeOr'] == swath]
    dates = df_swath['date'].unique()
    
    for date in dates:
        df_swath_date = df_swath[df_swath['date'] == date]
        df_swath_date.to_file(f'data/processed/parings/s1_{swath}_{date}.shp', driver='ESRI Shapefile')

    
    
```
